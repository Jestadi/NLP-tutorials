## Tokenizing by scraping a web source - Task 4.1

```python


import nltk
from nltk.tokenize import word_tokenize
from urllib import request
```

```python
url = "https://www.gutenberg.org/files/98/98-0.txt"
```

```python
response = request.urlopen(url)
```

```python
raw = response.read().decode('utf8')
```

```python
tokens = word_tokenize(raw)
```

## Stemmers - Task 4.2

```python
#different types of stemmers 
from nltk.stem import PorterStemmer
porter = PorterStemmer()
print(porter.stem('joyous'))
print(porter.stem('happiness'))
print(porter.stem('cacti'))
```

```python
from nltk.stem import LancasterStemmer
lancaster = LancasterStemmer()
print(lancaster.stem('joyous'))
print(lancaster.stem('happiness'))
print(lancaster.stem('cacti'))
print(lancaster.stem('singing'))
```

```python
from nltk.stem import RegexpStemmer
regex = RegexpStemmer('ing$')
print(regex.stem('joyous'))
print(regex.stem('happiness'))
print(regex.stem('cacti'))
print(regex.stem('singing'))
```

```python
from nltk.stem import SnowballStemmer
snow = SnowballStemmer('spanish')
snow.stem('encontramos')

```

```python
#Stemming before vectorization - reduces complexity while computing

stem = []
for i in tokens:
    stem.append(lancaster.stem(i))
stem
```
