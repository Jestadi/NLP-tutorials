## Named Entity Recognition

```python
import spacy
from spacy.lang.en.examples import sentences 
m = spacy.load('en_core_web_sm')
```

```python
from bs4 import BeautifulSoup
```

```python
import requests 
import re
```

```python
URL = "https://www.zeebiz.com/market-news/news-deutsche-bank-shares-drop-amid-global-jitters-over-banks-227493"
```

```python
html_content = requests.get(URL).text
soup = BeautifulSoup(html_content, "lxml")
```

```python
body = soup.body.text
```

```python
print(body)
```

```python
#body[1000:1500]
#Getting text between two substring because we are only interested in the article

test_str = body
 
# initializing substrings
sub1 = "Markets News"
sub2 = "Get Latest Business News"
 
# getting index of substrings
idx1 = test_str.index(sub1)
idx2 = test_str.index(sub2)
 
res = ''
# getting elements in between
for idx in range(idx1 + len(sub1) + 1, idx2):
    res = res + test_str[idx]
 
# printing result
print("The extracted text : " + res)
```
